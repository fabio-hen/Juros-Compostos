```python
from math import pow

def juros_simples (capital_inicial, taxa_juros, tempo):
    montante_simples = capital_inicial * (1 + taxa_juros * tempo)   #M = C * (1 + i * t)
    return montante_simples


def composto (capital_inicial, taxa_juros, tempo):
    montante = capital_inicial * pow((1 + taxa_juros), tempo)   #M = C * (1 + i)^t
    return montante

capital = float(input("Digite o capital inicial: "))
Juros = float(input("Digite a taxa de juros anual em porcentagem: "))
Juros = Juros / 100 #converte porcentagem para decimal
tempo = int(input("Digite o tempo em meses: "))
tempo = tempo / 12  #converte meses para anos

montante = composto(capital, Juros, tempo)
print(f"O montante após {tempo*12:.0f} meses será de R$ {montante:.2f}")
print(f"O juros compostos ganho será de R$ {montante - capital:.2f}")

montante_simples = juros_simples(capital, Juros, tempo)
print(f"O montante após {tempo*12:.0f} meses será de R$ {montante_simples:.2f}")
print(f"O juros simples ganho será de R$ {montante_simples - capital:.2f}")
```
